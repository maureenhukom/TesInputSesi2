```python
print("MENU MENGHITUNG RUMUS LUAS SEGITIGA DAN PERSEGI")
print("Masukkan angka 1 untuk menghitung luas segitiga")
print("Masukkan angka 2 untuk menghitung luas persegi")
print("Pilihan menu yang diinginkan : ")
angka = int(input())

if angka == 1:
    print("Masukkan tinggi segitiga: ")
    tinggi_segitiga = int(input())
    print("Masukkan alas segitiga: ")
    alas_segitiga = int(input())
    a = (tinggi_segitiga*alas_segitiga)/2
    print(a)
elif angka == 2:
    print("Masukkan sisi persegi: ")
    sisi_persegi = int(input())
    a = sisi_persegi*sisi_persegi
    print(a)
else :
    print("Masukkan input yang benar")
```
